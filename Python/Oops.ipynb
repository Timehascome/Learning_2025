```python
#Inheritance #Polymorphism #Encapsulation #Abstraction
class Animal:
    def __init__(self,name):
        self.name = name

    def voice(self):
        if self.name == "Dog":
           return "Bark"
        elif self.name == "Cat":
            return "Meow"
        else:
            return "NA"

class Dog(Animal):
      def __init__(self, breed):
          self.breed = breed
          print(breed)
          ins = Animal("Dog")
          sound = ins.voice()
          print(sound)                            

## In the above example we inherited the class Animal and used the method voice
         
x = Dog("Alsetian")
print(x)

```

```python
# Abstraction
from abc import ABC, abstractmethod
import math
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
        
class Circle(Shape):
    def __init__(self,radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius**2
        
circle = Circle(5)
print(circle.area())
```

```python
#polymorphism

class Dog:
    def make_sound(self):
        print("Woof")

class Cat:
    def make_sound(self):
        print("Meow")

a = Dog()
b = Cat()
a.make_sound()
b.make_sound()
    
```

```python
#Encapsulation

class banking:
    def __init__(self,balance):
        self._balance = balance
    def deposit(self, amount):
        self._balance+= amount
    def get_balance(self):
        return self._balance

x = banking(100)
y = x.deposit(100)
balance = x.get_balance()
print(balance)

        
        
        
```
