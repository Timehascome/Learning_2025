```python
import pandas as pd

list_of_numbers = [x for x in range(0,5)]
print(len(list_of_numbers))

#Series can be used when data is one dimesional and DataFrame in pandas can be used for two dimensional 
df_Series = pd.Series(list_of_numbers)
df_DataFrame = pd.DataFrame(list_of_numbers)
print(f"The series is single column style data as below .\n{df_Series}, The DataFrame is row x columns format as below .\n{df_DataFrame}")

#Use name ='Column_name' for Series and index = [] for custom values
#Use column =[] and index =[] to assign column and index for custom values while creation of DataFrame
custom_columns = ["col_1"]
index_custom = ["row_"+str(i) for i in range(len(list_of_numbers))]

df_Series = pd.Series(list_of_numbers, name= 'Col_1', index=index_custom )  #pass directly as single column
print(f"Series with custom column name and indices .\n{df_Series}")

df_DataFrame = pd.DataFrame(list_of_numbers, columns=custom_columns, index=index_custom)  #pass the list of columns
print(f"DataFrame with custom columns and indices .\n{df_DataFrame}")


#Lets Load a DataFrame from a csv file
df_customers_sample = pd.read_csv("../customers.csv", nrows=10) #good practice to intially load 10 lines to see the data nrows help with that
print(df_customers_sample.head())
#Lets check the memory and optimize for processing
df_customers = pd.read_csv("../customers.csv")
df_customers.info(memory_usage='deep')  #shows the data type and memory in this case 45 MB
#Lets create a copy and see if we can reduce the memory by changing data types
#from object to Category can be applied to columns with repeating values say gender 
df_customers_copy = df_customers.copy()
df_customers_copy["gender"] = df_customers_copy["gender"].astype("category")
df_customers_copy.info(memory_usage='deep') #shows memory reduced to 40 MB


#THE ABOVE IS A GOOD STRATEGY TO PREPARE DATA WITH USER FRIENDLY COLUMNS, INDICES AND COMPUTE FRIENDLY MEMORY REDUCTION.

```
